# Query (Consulta) de dados no DataFrame

```python
import pandas as pd
from pandas import DataFrame
```

```python
df = pd.read_csv("dataset2.csv")
```

```python
df.head(5)
```

```python
# Checamos os valores mínimo e máximo da coluna Valor_Venda
df.Valor_Venda.describe()
```

O intervalo de vendas (Valor_Venda) é de 0.44 a 22638. Vamos fazer uma consulta e retornar todas as vendas dentro de um range de valores.

```python
# Geramos um novo dataframe apenas com o intervalo de vendas entre 229 e 1000
df2 = df.query('229 < Valor_Venda < 10000')
```

```python
# Então confirmamos os valores mínimo e máximo 
df2.Valor_Venda.describe()
```

```python
# Geramos um novo dataFrame apenas com os valores de venda acima da média
df3 = df2.query('Valor_Venda > 766')
```

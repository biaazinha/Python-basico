# Filtrando DataFrame Do Pandas com base em Strins

```python
import pandas as pd
from pandas import DataFrame
df = pd.read_csv("dataset2.csv")
```

```python
df.head()
```

```python
df.Segmento.value_counts()
```

```python
# Filtramos o DataFrame pela coluna Segmento com valores que iniciam com as letras 'Con'
df[df.Segmento.str.startswith('Con')].head()
```

```python
# Filtramos o DataFrame pela coluna Segmento com valores que terminam com as letras 'mer'
df[df.Segmento.str.endswith('mer')].head()
```

# Split de Strings em DataFrames 

```python
df['ID_Pedido'].head() # País, ano e id do pedido
```

```python
# Split da coluna pelo caracter '-'
df['ID_Pedido'].str.split('-')
```

```python
df['ID_Pedido'].str.split('-').str[1].head() # Extraindo o ano
```

```python
# Fazemos o split da coluna e extraímos o item na posição 2 (indice 1)
df['Ano'] = df['ID_Pedido'].str.split('-').str[1]
```

```python
# Conferimos a nova coluna criada
df.head()
```

# Strip de Strings em DataFrames

```python
df.head(3)
```

```python
df['Data_Pedido'].head(3)
```

```python
# Vamos remover os dígitos 2 e 0 á esquerda do valor da variável 'Data_Pedido'
df['Data_Pedido'].str.lstrip('20') # Removendo os caracteres 20
```

# Replace de Strings (substituir)

```python
df.head()
```

```python
# Substituir CG por AX
df['ID_Cliente'] = df['ID_Cliente'].str.replace('CG', 'AX')
```

```python
df.head()
```

# Combinações de strings

```python
# Concatenando strings 
df['Pedido_Segmento'] = df['ID_Pedido'].str.cat(df['Segmento'], sep = '-')
```

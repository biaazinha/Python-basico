# Gráfico de Dispersão (Pontos)

```python
import matplotlib .pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,5,6,8,4,8]
```

```python
plt.scatter(x, y, label = 'Pontos', color = 'red', marker = '*')
plt.legend()
plt.show()
```

# Gráfico de Área Empilhada

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
passear = [8,5,7,8,13]
```

```python
plt.stackplot(dias, dormir, comer, trabalhar, passear, colors = ['m', 'c', 'r', 'k', 'b'])
plt.show()
```

# Gráfico de Pizza

```python
fatias = [7,2,2,13]
atividades = ['dormir', 'comer', 'passear', 'trabalhar']
cores = ['olive', 'lime', 'violet', 'royalblue']
```

```python
plt.pie(fatias, labels = atividades, colors = cores, startangle = 90, shadow = True, explode = (0,0.2,0,0))
plt.show()
```
